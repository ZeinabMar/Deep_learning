```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
```

```python
x = np.linspace(-10,10,100)
```

```python
def Relu(x):
  y = np.where(x<=0,0,x)
  return y

def Drv_Relu(x):
  y = np.where(x<0,0,1)
  return  y
```

```python
def Elu(x,alpha):
  y=np.where(x<0,alpha*(np.exp(x)-1),x)
  return y

def Drv_Elu(x,alpha):
  y = np.where(x<0, alpha*np.exp(x), 1)
  return y
```

```python
def Tanh(x):
  y =  (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
  return y

def Drv_Tanh(x):
  gx =  (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
  y = 1-np.power(gx,2)
  return y
```

```python
def Sigmoid(x):
  y =  1/(1+np.exp(-x))
  return y

def Drv_Sigmoid(x):
  y = (1+np.exp(-x)+(x*np.exp(-x)))/np.power(1+np.exp(-x),2)
  return y
```

```python
def leaky_relu(x, alpha=0.1):
  y =  np.where(x<=0,0.01*x,x)
  return y

def Drv_leaky_relu(x,alpha=0.1):
  y = np.where(x<=0,0.01,1)
  return y
```

```python
def Softmax(x):
  e_x = np.exp(x-np.max(x))
  y = e_x/np.sum(e_x,axis=0)
  return y
```

### Plot Sigmoid

```python
y = Softmax(np.array([1,3,4]))
plt.xlabel("Classes")
plt.ylabel("Probability")
plt.title("Softmax")
plt.bar(['A','B','C'],y)
```

### Plot Tanh

```python
y = Tanh(x)
plt.xlabel("input")
plt.ylabel("output")
plt.plot(x,y)
#*********************
y_d = Drv_Tanh(x)
plt.grid(True)
plt.plot(x,y_d)
plt.legend(['Tanh','Deriviation_Tanh'])
```

### Plot Sigmoid

```python
y = Sigmoid(x)
plt.xlabel("input")
plt.ylabel("output")
plt.plot(x,y)
#*************************
y_d = Drv_Sigmoid(x)
plt.grid(True)
plt.plot(x,y_d)
plt.legend(['Sigmoid','Deriviation_Sigmoid'])
```

### Plot Leaky

```python
y = leaky_relu(x)
plt.xlabel("input")
plt.ylabel("output")
plt.plot(x,y)

#******************
y_d = Drv_leaky_relu(x)
plt.grid(True)
plt.plot(x,y_d)
plt.legend(['leaky_relu','Deriviation_leaky_relu'])
```

### Plot Elu


```python
y = Elu(x,0.5)
plt.xlabel("input")
plt.ylabel("output")
plt.plot(x,y)
#******************
y_d = Drv_Elu(x,0.5)
plt.grid(True)
plt.plot(x,y_d)
plt.legend(['Elu','Deriviation_Elu'])
```
